```python
import pandas as pede

df = pede.read_csv('titanic.csv')

print()
print('--- Ini adalah hasil dari Imputasi data dan menampilkan tujuh urutan awal pada data. ---\n')

mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age)

df['age'] = df['age'].astype(int)
df.head(7)
```
