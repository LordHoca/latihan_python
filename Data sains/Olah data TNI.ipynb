```python
import pandas as pede
df = pede.read_csv('datatni.csv')
df.drop('kategori', axis=1, inplace=True)
```

```python
print('--- Data Calon TNI ---')
print(df)
print()
print('--- Statistik Dasar Tinggi Badan ---')
rata_tinggi = df['tinggi'].mean()
tertinggi = df['tinggi'].max()
terendah = df['tinggi'].min()
median = df['tinggi'].median()
print(df['tinggi'].describe().round(3))
print()
print(f'Rata -rata tinggi        : {rata_tinggi} cm')
print(f'Tertinggi                : {tertinggi} cm')
print(f'Terdendah                : {terendah} cm')
print(f'Median atau Nilai tengah : {median} cm')
```

```python
print('--- Calon diatas rata - rata tinggi ---\n')
rata_tinggi = df['tinggi'].mean()
diatas_rata_tinggi = df[df['tinggi'] > rata_tinggi]
print(f'{diatas_rata_tinggi}\n')
print('--- Urutan dari tertinggi ke terendah ---\n')
urutan = df.sort_values(by='tinggi', ascending=False)
print(f'{urutan}\n')
df = pede.read_csv('datatni.csv')
print('--- Data dengan kategori ---\n')
print(df)
print('\n--- Jumlah calon berdasarkan Tinggi ---')
kategories = df['tinggi'].value_counts()
print(f'{kategories}\n')
print('\n--- Jumlah calon berdasarkan Kategori ---')
ciri = df['kategori'].value_counts()
print(ciri)
```

```python
print('--- Statistik Lanjutan ---\n')
std = round(df['tinggi'].std(), 2)
varsi = round(df['tinggi'].var(), 2)
print(f'Standar Deviasi : {std}')
print(f'Variansi        : {varsi}')
print('\n--- Calon dengan tinggi badan tertinggi ---\n')
data_tertinggi = df[df['tinggi'] == df['tinggi'].max()]
print(data_tertinggi)
```
